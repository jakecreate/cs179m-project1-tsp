```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
center_x, center_y = 1, 1
radius = 1

theta = np.random.uniform(0, 2 * np.pi, 50)

x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

fig, ax = plt.subplots()
ax.scatter(x, y, color='red')
ax.set_aspect('equal')
plt.show()
```

```python
# np.savetxt(f'../data/circle.txt', np.vstack([x,y]).T)
min_dist = 2*np.pi
```

```python
X_sa = np.loadtxt('../experiment/128Circle201_EXP_861.txt')
X_sa2 = np.loadtxt('../experiment1/128Circle201_EXP_631.txt')
X_sa3 = np.loadtxt('../experiment1/128Circle201_EXP_775.txt')
X_sa4 = np.loadtxt('../experiment1/128Circle201_EXP_795.txt')
X_sa5 = np.loadtxt('../experiment1/128Circle201_EXP_817.txt')

X_rs = np.loadtxt('../experiment/128Circle201_RS_4088.txt')

sam_100 = X_sa[:, 0] <= 100
rsm_100 = X_rs[:, 0] <= 100
X_sa = X_sa[sam_100]
X_rs = X_rs[rsm_100]

```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(np.arange(0, 100), np.full((100,),201), c='red', label='Ground Truth')
ax.plot(X_sa[:, 0], X_sa[:, 1], c='orange', alpha=0.3)
ax.plot(X_sa2[:, 0], X_sa2[:, 1], c='orange', label='Simulated Annealing (Best Outcome)')
ax.plot(X_sa3[:, 0], X_sa3[:, 1], c='orange', alpha=0.3)
ax.plot(X_sa4[:, 0], X_sa4[:, 1], c='orange', alpha=0.3)
ax.plot(X_sa5[:, 0], X_sa5[:, 1], c='orange', alpha=0.3)

ax.plot(X_rs[:, 0], X_rs[:, 1], c='blue', label='Random Search')
ax.set_ylabel('Distance (meters)')
ax.set_xlabel('Time (seconds)')
ax.set_title('128Circle201 Benchmark Test Results')
plt.legend()
```
