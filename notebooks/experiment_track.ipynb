```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
center_x, center_y = 1, 1
radius = 1

theta = np.random.uniform(0, 2 * np.pi, 50)

x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

fig, ax = plt.subplots()
ax.scatter(x, y, color='red')
ax.set_aspect('equal')
plt.show()
```

```python
# np.savetxt(f'../data/circle.txt', np.vstack([x,y]).T)
min_dist = 2*np.pi
```

```python
X_sa = np.loadtxt('../experiment/128Circle201.txt_EXP_861.txt')
X_rs = np.loadtxt('../experiment/128Circle201.txt_EXP_4088.txt')

sam_100 = X_sa[:, 0] <= 100
rsm_100 = X_rs[:, 0] <= 100
X_sa = X_sa[sam_100]
X_rs = X_rs[rsm_100]

```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(np.full(int(np.max(X_sa[:, 0])), 201), label='Ground Truth')
ax.plot(X_sa[:, 0], X_sa[:, 1], label='Simulated Annealing')
ax.plot(X_rs[:, 0], X_rs[:, 1], label='Random Search')
ax.set_ylabel('Distance (meters)')
ax.set_xlabel('Time (seconds)')
ax.set_title('128Circle201 Data Test')
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(np.arange(0, 100), 201, label='ground truth')
ax.plot(X_sa[:, 0], X_sa[:, 1], label='simulated annealing')
ax.plot(X_rs[:, 0], X_rs[:, 1], label='random search')

plt.legend()
```
