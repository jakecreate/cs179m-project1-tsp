```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)
x = np.random.uniform(5 - np.sqrt(5), 5 + np.sqrt(5), 15)
y = np.sqrt(5 - (x - 5)**2) + 5

x = np.vstack([x.reshape(-1, 1), x.reshape(-1, 1)]).ravel()
y = np.vstack([y.reshape(-1,1), -y.reshape(-1,1)]).ravel()

X = np.vstack([x, y]).T
X[:5]
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X[:, 0], X[:, 1])
for i, (x_i, y_i) in enumerate(X):
    ax.annotate(f'{i}', xy=(x_i, y_i), color='black', va='bottom', ha='left', fontsize='small')

plt.plot()
```
